```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv("nodes.csv", header=None)
```

```python
for i in range(1000):
    if len(eval(data_df[0].get(i)))!=100:
        print(i)
```

```python
data_df = data_df.drop(356)
```

```python
X = [eval(data_df.iloc[i][0]) for i in range(len(data_df))]
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0, perplexity=30)
```

```python
tsne_obj= tsne.fit_transform(X)
```

```python
import seaborn as sns
```

```python
tsne_df = pd.DataFrame({'X':tsne_obj[:,0],
                        'Y':tsne_obj[:,1],
                        'Cluster':data_df[1]})

sns_plot = sns.scatterplot(x="X", y="Y",
              hue="Cluster",
              palette=['purple','red','brown','blue','green'],
              legend='full',
              data=tsne_df).set_title('Tweets Clustering');
```

```python
fig = sns_plot.get_figure()
```

```python
fig.savefig('plot.png', dpi=400)
```
